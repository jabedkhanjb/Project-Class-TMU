# Matplotlib Data Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([1, 2, 3, 4, 5])

y1 = np.array([5, 8, 5, 14, 7])
y2 = np.array([2, 6, 3, 9, 12])

design = plt.figure(figsize = (8, 5), facecolor = "g", edgecolor = "c", linewidth = 5)
mydesign = design.add_axes([1,1,1,1])


mydesign.plot(x1, y1)
mydesign.plot(x2, y2)


# All labels and titles

plt.legend(labels = (["Graph 1", "Graph 2"]))
plt.show()
```
