# Setting Limits and Tick labels

<b>xlim()</b> and <b>ylim()</b> functions are used to set the limits of the X-axis and Y-axis respectively. Similarly, <b>xticks()</b> and <b>yticks()</b> functions are used to set tick labels.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([10, 15, 20, 30, 35, 40])
ypoint = np.array([20, 25, 30, 50, 52, 35])

plt.plot(xpoint, ypoint, "h-k", ms="15", mfc="m", mec="c")

plt.title("Project Class Data Analysis", fontsize = "20", color = "#062a4c")
plt.xlabel("X-Axis Data", fontsize = 12, color="#859105")
plt.ylabel("Y-Axis Data", fontsize = 12, color="#859105")

# setting the limit of yaxis
plt.ylim(15, 100)
plt.xlim(8, 42)

# setting the label of xaxis
plt.xticks(xpoint, labels=["one", "two", "three", "four", "five", "six"])
# plt.yticks(ypoint, labels=["one", "two", "three", "four", "five", "six"])


plt.show()
```

# Adding Legends

<b>Syntax: </b> <i>matplotlib.pyplot.legend([“name1”, “name2”], bbox_to_anchor=(x, y), ncol=1)</i>

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([10, 15, 20, 30, 35, 40])
ypoint = np.array([20, 25, 30, 50, 52, 35])

plt.plot(xpoint, ypoint, "h-k", ms="15", mfc="m", mec="c")

plt.title("Project Class Data Analysis", fontsize = "20", color = "#062a4c")
plt.xlabel("X-Axis Data", fontsize = 12, color="#859105")
plt.ylabel("Y-Axis Data", fontsize = 12, color="#859105")

# setting the limit of yaxis
plt.ylim(15, 100)
plt.xlim(8, 42)

# setting the label of xaxis
plt.xticks(xpoint, labels=["one", "two", "three", "four", "five", "six"])
# plt.yticks(ypoint, labels=["one", "two", "three", "four", "five", "six"])

# adding legend
plt.legend(["Linear Graph"])


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

x1 = np.array([1, 2, 3, 4, 5])
x2 = np.array([1, 2, 7, 4, 5])

y1 = np.array([5, 8, 5, 14, 7])
y2 = np.array([2, 6, 3, 9, 12])

design = plt.figure(figsize = (10, 5), facecolor = "g", edgecolor = "c", linewidth = 5)
mydesign = design.add_axes([1,1,1,1])


mydesign.plot(x1, y1, "h-m", ms = 15, mec = "m", mfc = "c")
mydesign.plot(x2, y2, "h-y", ms = 15, mec = "y", mfc = "k")


# All labels and titles
plt.title("Data Visualization", fontsize = 25)
plt.xlabel("X-Axis", fontsize = 20, color = "white")
plt.ylabel("Y-Axis", fontsize = 20, color = "white")

plt.xlim(0, 8)
plt.ylim(0, 20)
# plt.xticks(x1, labels=["One", "Two", "Three", "Four", "Five"])
plt.legend(labels = (["Graph 1", "Graph 2"]))
plt.show()
```
