```python
# mean, median, mode
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# find mean
meanvalue = np.mean(speed)
print(meanvalue)

plt.plot(speed, marker = "o", color = "c")
plt.plot(meanvalue, color = "m", marker = "o")
plt.grid()
plt.show()
```

```python
# find median
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.figure import Figure

speed = [99,86,48,88,111,86,95,103,87,94,78,77,85,86]
print(sorted(speed))
medianvalue = np.median(speed)

print(medianvalue)
plt.plot(medianvalue, marker = "o")
plt.show()
```

```python
# mode()

from scipy import stats

speed = [99,89,86,87,88,111,89,86,103,87,89,89,94,78,77,85,86]

x = stats.mode(speed)
print(x)
```
