# Finding Variance Manually 

```python
# speed = [86,157,88,86,127,25,286]
def sum_(speed):
    sum = 0
    
    for i in speed:
        sum = sum + i
        
    return sum

# speed = []
speed = [86,157,88,86,127,25,286]

total = sum_(speed)

# print(f"Addition of {speed} is |*'{total}'*|")

arraylen = len(speed)

mean = total / arraylen

var = []
for i in speed:
    myarray = i - mean
    var.append(myarray)
    
# print(var)

rootvar = []

for i in var:
    squre = i * i
    rootvar.append(squre)
    
# print(rootvar)

sumroot = 0
for i in rootvar:
    sumroot = sumroot + i
    
variance = sumroot / arraylen
print(variance)
```
